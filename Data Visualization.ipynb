# Data Visualization Techniques

## Seaborn

Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.




- **Trends** - A trend is defined as a pattern of change.
sns.lineplot
    - **Line charts** are best to show trends over a period of time, and multiple lines can be used to show trends in more than one group.
- **Relationship** - There are many different chart types that you can use to understand relationships between variables in your data.
    - sns.barplot - Bar charts are useful for comparing quantities corresponding to different groups.
    - `sns.heatmap` - Heatmaps can be used to find color-coded patterns in tables of numbers.
    - `sns.scatterplot` - Scatter plots show the relationship between two continuous variables; if
    - `color-coded`, we can also show the relationship with a third categorical variable.
    - `sns.regplot` - Including a regression line in the scatter plot makes it easier to see any linear relationship between two variables.
    - `sns.lmplot` - This command is useful for drawing multiple regression lines, if the scatter plot contains multiple, color-coded groups.
    - `sns.swarmplot` - Categorical scatter plots show the relationship between a continuous variable and a categorical variable.
- Distribution - We visualize distributions to show the possible values that we can expect to see in a variable, along with how likely they are.
    - `sns.histplot` - Histograms show the distribution of a single numerical variable.
    - `sns.kdeplot` - KDE plots (or 2D KDE plots) show an estimated, smooth distribution of a single numerical variable (or two numerical variables).
    - `sns.jointplot` - This command is useful for simultaneously displaying a 2D KDE plot with the corresponding KDE plots for each individual variable.

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")

df = pd.read_csv("./data/spotify.csv", parse_dates=True, index_col="Date")
df.head()
```

### Lineplot

```python
plt.figure(figsize=(8, 4))  # set the wigth and height in inches

plt.title("Daily Global Streams of Popular Songs in 2017-2018")  # chart name

plt.xlabel("Date")  # label X axis

sns.lineplot(data=df)  # plot with all data
```

```python
sns.lineplot(data=df['Shape of You'], label="Shape of You")  # plot 1 col
sns.lineplot(data=df['Despacito'], label="Despacito")        # add another line to plot
```

### Bar Chart

```python
# Import data
df_flight = pd.read_csv("./data/flight_delays.csv")

df_flight.head()
```

```python
plt.figure(figsize=(12, 8))

plt.title("Average Arrival Delay")

sns.barplot(data=df_flight)
```

```python
plt.title("Average Arrival Delay of AA Airline")

sns.barplot(x=df_flight.index, y=df_flight['AA'])

plt.ylabel("Araival delay in minutes")
plt.xlabel("Month")
```

### Heatmap

A heatmap is a data visualization technique used to represent the magnitude of values in a matrix or grid format, where individual values are represented by different colors. The purpose of a heatmap is to provide a graphical representation of data where color intensity or gradient visually indicates the value's magnitude.

And use used for Visual Pattern Recognition. Heatmaps make it easier to **identify patterns, correlations**, and outliers in large datasets through color gradients.

```python
# Set the width and height of the figure
plt.figure(figsize=(14,7))

plt.title("Average Arrival Delay for Each Airline, by Month")

# Heatmap showing average arrival delay for each airline by month
sns.heatmap(data=df_flight, annot=True)
# annot=True -- This ensures that the values for each cell appear on the chart. 
#               (Leaving this out removes the numbers from each of the cells!)


plt.xlabel("Airline")
```

### Scatter Plots

```python
insurance = pd.read_csv("./data/insurance.csv")
print(insurance.head())
print(insurance.describe())
```

```python
plt.title("Insurance Body Mass Index (BMI) to Charges")
plt.xlabel("BMI")
plt.ylabel("Charges")
sns.set_style("dark")
sns.scatterplot(x=insurance['bmi'], y=insurance['charges'])
```

#### Regression Scatter Plot

```python
sns.regplot(x=insurance['bmi'], y=insurance['charges'])
```

#### Color-coded scatter plot

```python
sns.scatterplot(x=insurance['bmi'], y=insurance['charges'], hue=insurance['smoker'])
```

#### Multilinear Regression scatter plot

```python
sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance)
```

#### Swarmplot

```python
sns.swarmplot(x=insurance['smoker'], y=insurance['charges'])
```

### Histogram 

```python
iris_filepath = "./data/iris.csv"

df_iris = pd.read_csv(iris_filepath, index_col="Id") 
df_iris.head()
```

```python
sns.histplot(df_iris['Petal Length (cm)'])
```

#### Color-coded plots

```python
sns.histplot(data=df_iris, x='Petal Length (cm)', hue='Species')
```

### Density plots

#### Kernel density estimate (KDE)

```python
sns.kdeplot(data=df_iris['Petal Length (cm)'], fill=True)
```

```python
sns.kdeplot(data=df_iris, x='Petal Length (cm)', hue='Species', fill=True)
```

#### 2D KDE plots 

A 2D KDE plot shows the density of data points in two dimensions. It uses a smooth surface or contour lines to represent areas where data points are more or less concentrated. This helps visualize the distribution, clusters, and relationships between two variables.

```python
sns.jointplot(x=df_iris['Petal Length (cm)'], y=df_iris['Sepal Width (cm)'], kind="kde")
sns.set_style("whitegrid")

# "darkgrid", (2)"whitegrid", (3)"dark", (4)"white", and (5)"ticks"
```
