# Data Visualization Techniques

## Seaborn

Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.




```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")

df = pd.read_csv("./data/spotify.csv", parse_dates=True, index_col="Date")
df.head()
```

### Lineplot

```python
plt.figure(figsize=(8, 4))  # set the wigth and height in inches

plt.title("Daily Global Streams of Popular Songs in 2017-2018")  # chart name

plt.xlabel("Date")  # label X axis

sns.lineplot(data=df)  # plot with all data
```

```python
sns.lineplot(data=df['Shape of You'], label="Shape of You")  # plot 1 col
sns.lineplot(data=df['Despacito'], label="Despacito")        # add another line to plot
```

### Bar Chart

```python
# Import data
df_flight = pd.read_csv("./data/flight_delays.csv")

df_flight.head()
```

```python
plt.figure(figsize=(12, 8))

plt.title("Average Arrival Delay")

sns.barplot(data=df_flight)
```

```python
plt.title("Average Arrival Delay of AA Airline")

sns.barplot(x=df_flight.index, y=df_flight['AA'])

plt.ylabel("Araival delay in minutes")
plt.xlabel("Month")
```

### Heatmap

A heatmap is a data visualization technique used to represent the magnitude of values in a matrix or grid format, where individual values are represented by different colors. The purpose of a heatmap is to provide a graphical representation of data where color intensity or gradient visually indicates the value's magnitude.

And use used for Visual Pattern Recognition. Heatmaps make it easier to **identify patterns, correlations**, and outliers in large datasets through color gradients.

```python
# Set the width and height of the figure
plt.figure(figsize=(14,7))

plt.title("Average Arrival Delay for Each Airline, by Month")

# Heatmap showing average arrival delay for each airline by month
sns.heatmap(data=df_flight, annot=True)
# annot=True -- This ensures that the values for each cell appear on the chart. 
#               (Leaving this out removes the numbers from each of the cells!)


plt.xlabel("Airline")
```

### Scatter Plots

```python
insurance = pd.read_csv("./data/insurance.csv")
print(insurance.head())
print(insurance.describe())
```

```python
plt.title("Insurance Body Mass Index (BMI) to Charges")
plt.xlabel("BMI")
plt.ylabel("Charges")

sns.scatterplot(x=insurance['bmi'], y=insurance['charges'])
```

#### Regression Scatter Plot

```python
sns.regplot(x=insurance['bmi'], y=insurance['charges'])
```

#### Color-coded scatter plot

```python
sns.scatterplot(x=insurance['bmi'], y=insurance['charges'], hue=insurance['smoker'])
```

#### Multilinear Regression scatter plot

```python
sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance)
```

#### Swarmplot

```python
sns.swarmplot(x=insurance['smoker'], y=insurance['charges'])
```
