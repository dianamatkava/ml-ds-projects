# Inverse matrix

An inverse matrix is a matrix that, when multiplied by a given square matrix, *yields the identity matrix*. Here’s a detailed overview:

### Definition

For a square matrix $A$, its inverse is denoted as $A^{-1}$. The relationship is defined as:

$A \cdot A^{-1} = A^{-1} \cdot A = I$

where $I$ is the [[I - Identity Matrix]] of the same size as $A$.

### Conditions for Inverses

1. **Square Matrix**: Only square matrices (same number of rows and columns) can have inverses.
2. **Invertibility**: A matrix $A$ is invertible (or non-singular) if there exists a matrix $A^{-1}$ such that the above condition holds. If no such matrix exists, $A$ is called singular.
3. **Determinant**: For a matrix to be invertible, its determinant must be non-zero. If the determinant of $A$ (denoted as $\text{det}(A)$ is zero, the matrix does not have an inverse.

### How to Find the Inverse

There are several methods to find the inverse of a matrix, including:

1. **Gauss-Jordan Elimination**: You can augment the matrix $A$ with the identity matrix $I$ and use row operations to transform $A$ into III. The resulting matrix on the right will be $A^{-1}$.
2. **Adjugate and Determinant Method**: For a $2 \times 2$ matrix:
    
    $A = \begin{pmatrix} a & b \\ c & d \end{pmatrix}$
    
    The inverse is given by:
    
    $A^{-1} = \frac{1}{ad - bc} \begin{pmatrix} d & -b \\ -c & a \end{pmatrix}$
    
    provided ad-bc != 0
    
	The determinant of a $3 \times 3$ matrix $E = \begin{pmatrix} a & b & c \\ d & e & f \\ g & h & i \end{pmatrix}$​​ is calculated as:
	
	$\text{det}(E) = a(ei - fh) - b(di - fg) + c(dh - eg)$

### Properties of Inverse Matrices

1. **Unique**: If a matrix has an inverse, it is unique.
2. **Inverse of a Product**: The inverse of a product of two matrices is the product of their inverses in reverse order:
    
    $(AB)^{-1} = B^{-1}A^{-1}$
3. **Inverse of the Inverse**: The inverse of an inverse matrix is the original matrix:
    
    $(A^{-1})^{-1} = A$
4. **Transpose**: The inverse of the transpose of a matrix is equal to the transpose of the inverse:
    
    $(A^T)^{-1} = (A^{-1})^T$


# Identity Matrix 
The identity matrix is a special type of square matrix that plays a crucial role in linear algebra. Here are the key characteristics and meanings of the identity matrix:

### Definition

1. **Square Matrix**: The identity matrix is always a square matrix, meaning it has the same number of rows and columns.
2. **Diagonal Elements**: The elements on the main diagonal (from the top left to the bottom right) are all 1s.
3. **Off-Diagonal Elements**: All other elements (those not on the diagonal) are 0s.

### Notation

The identity matrix of size $n\times n$ is often denoted as $I_n$​. For example, the identity matrix of size #2 \times 2# is:

$I_2 = \begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix}$

And for size $3 \times 3$:

$I_3 = \begin{pmatrix} 1 & 0 & 0 \\ 0 & 1 & 0 \\ 0 & 0 & 1 \end{pmatrix}$

- *Leading value of Identity matrix is always 1* (row echelon form)

### Properties

1. **Multiplicative Identity**: The identity matrix acts as the multiplicative identity in matrix multiplication. For any matrix $A$ of compatible size:
    
    $A \cdot I = I \cdot A = A$
    
    This means that multiplying any matrix by the identity matrix does not change the matrix.
    
2. **Inverses**: If a matrix $A$ has an inverse $A^{-1}$, then:
    
    $A \cdot A^{-1} = I$
    
    This property is essential for solving systems of equations and understanding linear transformations.
    
3. **Linear Transformations**: In the context of linear transformations, the identity matrix represents the transformation that leaves vectors unchanged. For example, applying the transformation represented by $I$ to any vector $\mathbf{v}$ results in $\mathbf{v}$.


# Transpose of a Matrix

The **transpose of a matrix** is an operation that flips the matrix over its diagonal, turning the rows of the original matrix into columns and the columns into rows.

If you have a matrix $A$, the transpose of $A$ is denoted as $A^T$.

For example, if the original matrix $A$ is:
$$A = \begin{pmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \end{pmatrix}$$

The transpose ATA^TAT would be:
$$A^T = \begin{pmatrix} 1 & 4 \\ 2 & 5 \\ 3 & 6 \end{pmatrix}$$​​
So, in the transpose:
- The first row of $A$ becomes the first column of $A^T$.
- The second row of $A$ becomes the second column of $A^T$, and so on.

For an $n \times m$ matrix, its transpose will be an $m \times n$ matrix.
